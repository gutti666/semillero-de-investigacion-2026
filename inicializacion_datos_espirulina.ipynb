### 6.1. Distribución de Variables Principales

#### Variables de Color y Biomasa - Explicación Básica

**Variables de Biomasa:**
- **Porcentaje Verde (%)**: Proporción de píxeles verdes en la imagen. Mayor valor = mayor densidad de algas.
- **Absorbancia Estimada**: Medida de cuánta luz absorbe el cultivo. Valores altos (3-4) indican alta concentración de biomasa.

**Canales RGB (Red-Green-Blue):**
- **R_ROI**: Intensidad del color rojo (0-255). En Espirulina se absorbe por la clorofila → valores bajos indican cultivo denso.
- **G_ROI**: Intensidad del color verde (0-255). Es el canal DOMINANTE en Espirulina porque refleja luz verde → valores altos indican buena salud.
- **B_ROI**: Intensidad del color azul (0-255). Fuertemente absorbido por clorofila → valores muy bajos indican fotosíntesis activa.

**Canales HSV (Hue-Saturation-Value):**
- **H_ROI**: Matiz o tono del color (0-180 en OpenCV). Define el "tipo" de color: ~40-50 = verde amarillento normal en Espirulina.
- **S_ROI**: Saturación o pureza del color (0-255). Valores altos (>240) = color puro sin contaminantes.
- **V_ROI**: Valor o brillo (0-255). Es el máximo entre R, G y B. En Espirulina sana: V ≈ G (verde dominante).

💡 **Nota**: Las variables con sufijo `_Mascara` son iguales pero solo del cultivo segmentado (sin burbujas ni espuma).

## 7. Limpieza de Datos

Proceso de limpieza y tratamiento de valores anómalos en el dataset.


### 7.1. Manejo de Valores Cero en Máscaras

Cuando el algoritmo de segmentación falla (por burbujas, espuma o reflejos), las variables de Máscara toman valor 0. Estos valores no representan el cultivo real y deben ser tratados.

**Estrategia**: Rellenar valores 0 con el promedio de valores válidos (no-ceros) de cada variable.


```python
# Identificar columnas de Máscara
mask_columns = ['R_Mascara', 'G_Mascara', 'B_Mascara', 'H_Mascara', 'S_Mascara', 'V_Mascara','Porcentaje Verde (%)', 'Absorbancia Estimada']

print("=" * 70)
print(" LIMPIEZA DE DATOS: MANEJO DE VALORES CERO EN MÁSCARAS")
print("=" * 70)

# Analizar valores cero antes de la limpieza
print("\n ANTES DE LA LIMPIEZA:")
print("-" * 70)
for col in mask_columns:
    zeros_count = (df[col] == 0).sum()
    zeros_pct = (zeros_count / len(df)) * 100
    print(f"{col:15s}: {zeros_count:6d} valores cero ({zeros_pct:5.2f}%)")

total_rows = len(df)
rows_with_any_zero = (df[mask_columns] == 0).any(axis=1).sum()
print(f"\n Registros con al menos un cero: {rows_with_any_zero:,} de {total_rows:,} ({rows_with_any_zero/total_rows*100:.2f}%)")

# Crear copia del dataframe para limpieza
df_clean = df.copy()

# Rellenar valores 0 con el promedio de valores no-ceros
print("\nPROCESO DE LIMPIEZA:")
print("-" * 70)
for col in mask_columns:
    # Calcular promedio excluyendo ceros
    non_zero_mean = df_clean[df_clean[col] > 0][col].mean()
    zeros_mask = df_clean[col] == 0
    zeros_filled = zeros_mask.sum()
    
    if zeros_filled > 0:
        # Rellenar ceros con el promedio
        df_clean.loc[zeros_mask, col] = non_zero_mean
        print(f"{col:15s}: {zeros_filled:6d} valores rellenados con promedio = {non_zero_mean:.2f}")
    else:
        print(f"{col:15s}: Sin valores cero")

# Verificar limpieza
print("\nDESPUÉS DE LA LIMPIEZA:")
print("-" * 70)
for col in mask_columns:
    zeros_count = (df_clean[col] == 0).sum()
    print(f"{col:15s}: {zeros_count:6d} valores cero ({'✅ OK' if zeros_count == 0 else '⚠️ REVISAR'})")

print("\n" + "=" * 70)
print(f"Dataset limpio: {len(df_clean):,} registros con máscaras válidas")
print("=" * 70)
```

### 7.3. Detección de Datos Atípicos (Outliers)

Análisis mediante boxplots para identificar valores extremos en todas las variables del dataset.


```python
# Variables para análisis de outliers (todas las numéricas)
all_numeric_vars = [
    'Porcentaje Verde (%)', 'Absorbancia Estimada',
    'R_ROI', 'G_ROI', 'B_ROI', 'H_ROI', 'S_ROI', 'V_ROI',
    'R_Mascara', 'G_Mascara', 'B_Mascara', 'H_Mascara', 'S_Mascara', 'V_Mascara'
]

# Crear figura grande para todos los boxplots
fig, axes = plt.subplots(7, 2, figsize=(16, 22))
fig.suptitle('Análisis de Datos Atípicos - Todas las Variables', 
             fontsize=16, fontweight='bold', y=0.995)

axes = axes.flatten()

# Generar boxplot para cada variable
for idx, var in enumerate(all_numeric_vars):
    ax = axes[idx]
    
    # Boxplot con datos limpios
    bp = ax.boxplot(df_clean[var], vert=True, patch_artist=True,
                    boxprops=dict(facecolor='lightblue', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2),
                    whiskerprops=dict(color='blue', linewidth=1.5),
                    capprops=dict(color='blue', linewidth=1.5),
                    flierprops=dict(marker='o', markerfacecolor='red', markersize=3, alpha=0.3))
    
    # Calcular estadísticas de outliers
    Q1 = df_clean[var].quantile(0.25)
    Q3 = df_clean[var].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_clean[(df_clean[var] < lower_bound) | (df_clean[var] > upper_bound)][var]
    outliers_pct = (len(outliers) / len(df_clean)) * 100
    
    # Etiquetas y estadísticas
    ax.set_title(f'{var}\nOutliers: {len(outliers):,} ({outliers_pct:.2f}%)', 
                 fontsize=10, fontweight='bold')
    ax.set_ylabel('Valor', fontsize=9)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Añadir texto con estadísticas clave
    stats_text = f"Q1={Q1:.1f}\nMediana={df_clean[var].median():.1f}\nQ3={Q3:.1f}"
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes,
            fontsize=7, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
```

### 7.4. Resumen de Outliers

Tabla resumen con el conteo y porcentaje de datos atípicos por variable.


```python
# Tabla resumen de outliers

print("📊 RESUMEN DE DATOS ATÍPICOS (OUTLIERS) - MÉTODO IQR")
print(f"{'Variable':<25s} {'Q1':>10s} {'Mediana':>10s} {'Q3':>10s} {'IQR':>10s} {'Outliers':>12s} {'% Total':>10s}")


outliers_summary = []

for var in all_numeric_vars:
    Q1 = df_clean[var].quantile(0.25)
    Q3 = df_clean[var].quantile(0.75)
    IQR = Q3 - Q1
    median = df_clean[var].median()
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df_clean[(df_clean[var] < lower_bound) | (df_clean[var] > upper_bound)][var]
    outliers_count = len(outliers)
    outliers_pct = (outliers_count / len(df_clean)) * 100
    
    outliers_summary.append({
        'Variable': var,
        'Q1': Q1,
        'Mediana': median,
        'Q3': Q3,
        'IQR': IQR,
        'Outliers': outliers_count,
        'Porcentaje': outliers_pct
    })

    print(f"{var:<25s} {Q1:>10.2f} {median:>10.2f} {Q3:>10.2f} {IQR:>10.2f} {outliers_count:>12,} {outliers_pct:>9.2f}%")

print("=" * 100)

# Estadísticas generales
total_outliers = sum([x['Outliers'] for x in outliers_summary])
avg_outliers_pct = sum([x['Porcentaje'] for x in outliers_summary]) / len(outliers_summary)

print(f"\n Total de outliers detectados: {total_outliers:,}")


```
